```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
```

```python
densenet = load_model('densenet169.h5')
```

```python
densenet.summary()
```

## Loading a single image

```python
#Loading a fake image
my_img = image.load_img('00JEP4Z36Z.jpg',target_size=(224,224))
```

```python
my_img
```

### Transforming to array

```python
my_img_arr = image.img_to_array(my_img)
```

```python
my_img_arr.shape
```

#### Expanding axis as the model requires the image as a batch

```python
my_img_arr = np.expand_dims(my_img_arr,axis=0)
```

```python
my_img_arr.shape
```

## Prediction

```python
densenet.predict(my_img_arr)
```
